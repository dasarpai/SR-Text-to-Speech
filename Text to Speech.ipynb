```python
from gtts import gTTS
import os
```

```python
text ="My name is Hari Prasad Thapliyal, I live in Bangalore. I love Data science and Sanskrit"
```

```python
output = gTTS(text, lang="en", )
```

```python
output.save("output.mp3")
```

```python
os.system("start output.mp3")
```
